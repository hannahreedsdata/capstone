```python
genre_list_basics = ['pop','electronic','dance','rock','folk','country','blues','gospel','r-n-b','hip-hop','jazz','reggae','world-music','new-age','classical']

```

```python
new_df3 = df[df['genre'].isin(genre_list_basics)]
```

```python
# new_df2['genre'].value_counts().sort_values(ascending=False)
```

```python
flierprops = dict(marker='o', markersize=3)
```

### Danceability

```python
genres_ordered_by_danceability = list(new_df3.groupby('genre')['danceability']
                                      .median().sort_values(ascending=False).index)

plt.figure(figsize=(7,7))
sns.boxplot(x='danceability', y='genre', data=new_df3, order=genres_ordered_by_danceability,
            flierprops=flierprops);

plt.title('Danceability by Genre \n')

plt.xlabel('\n Score')
plt.ylabel('Genre \n ');
```

### Energy

```python
genres_ordered_by_energy = list(new_df3.groupby('genre')['energy']
                                      .median().sort_values(ascending=False).index)

# from https://stackoverflow.com/questions/43342564/flier-colors-in-boxplot-with-matplotlib
flierprops = dict(marker='o', markersize=3)
# , markerfacecolor='darkgray', markeredgecolor='darkgray'

plt.figure(figsize=(7,7))
sns.boxplot(x='energy', y='genre', 
            data=new_df3, 
            order=genres_ordered_by_energy,
            flierprops=flierprops);

plt.title('Energy by Genre \n')

plt.xlabel('\n Score')
plt.ylabel('Genre \n ');
```
